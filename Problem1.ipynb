Assignment: problem 1 - use BFS to find the k-th level friend of a user in a given social network

```python
from collections import deque

def FindFriends(G, User, K):

    visited, queue = set(), deque([User])
    visited.add(User)

    # Set levels of each user to 0
    levels = {user:0 for user in G}

    # Keep exploring as long as the queue is not empty
    while queue:

        # Dequeue a vertex from queue
        vertex = queue.popleft()
        vertex_level = levels[vertex]

        # If not visited, mark it as visited, increment its level, and enqueue it
        for neighbor in G[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                levels[neighbor] = vertex_level+1
                queue.append(neighbor)

    # Print users whose level matches the requested level
    print("The requested level friends of " + User + " are: ")
    for user in levels:
        if levels[user] == K:
            print(user)
```

```python
# Graph of above network
G = {

    'Pam':['Peter','Roger'],
    'Bob':['Rob','Pam','Richard'],
    'Roger':['Anna'],
    'Peter':['Amy'],
    'Amy':[],
    'Richard':[],
    'Anna':[],
    'Rob':[]
}

FindFriends(G,'Bob',3)
print("-----")
FindFriends(G,'Pam',1)
```
