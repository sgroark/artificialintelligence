**Assignment 2: Water Jug Puzzle**



- You have two jugs of water, a five litre jug and a three litre jug. You can fill and empty the jugs as much as you want. How can you get exactly four litres into the five litre jug?



**Start State:**
- [5L jug: 0L, 3L jug: 0L]
- [0,0]



**End State:**
- [5L jug: 4l, 3L jug: 3L]
- [4,3]

```python
class WaterJugPuzzle():

    # state model using two integers: 3L and 5L
    def __init__(self):
        self.threeL = 0
        self.fiveL = 0

    # checks if 3L jug is overflowing
    def is_threeL_overflow(self):
        if self.threeL >= 4:
            return True

    def is_threeL_full(self):
        if self.threeL == 3:
            return True

    # checks if 5L jug is overflowing
    def is_fiveL_overflow(self):
        if self.fiveL > 6:
            return True

    def is_fiveL_full(self):
        if self.fiveL == 5:
            return True


    def state_ok(self):
        return not (self.is_threeL_overflow() or self.is_fiveL_overflow())

    def save_state(self):
        self._threeL = self.threeL
        self._fiveL = self.fiveL

    def undo_state(self):
        self.threeL = self._threeL
        self.fiveL = self._fiveL

    # Rule 1: fill up 5L jug
    def R1_fill_5L_jug(self):

        self.save_state()

        if self.fiveL == 0:
            self.fiveL = 5
        else:
            raise ValueError('Cannot fill up a non-empty jug.')

    # Rule 2: transfer contents of 5L to 3L
    def R2_move_contents_5L_to_3L(self):
        self.save_state()
        fiveVol = self.fiveL


        while not self.is_threeL_full() and not fiveVol == 0:
            self.fiveL -= 1
            self.threeL += 1
            fiveVol -= 1


    # Rule 3: empty 3L jug

    def R3_empty_3L_jug(self):
      self.save_state()
      self.threeL = 0

    # print the state model

    def print_state(self):
      print("==> 5L Jug: {b}, 3L Jug: {l}".format(b=self.fiveL,l=self.threeL))


    # Rule 4:





############################


myobject = WaterJugPuzzle()

myobject.save_state()

myobject.print_state()
print('----------')

myobject.R1_fill_5L_jug()

myobject.print_state()

print('----------')
myobject.R2_move_contents_5L_to_3L()
myobject.print_state()

print('----------')
myobject.R3_empty_3L_jug()
myobject.print_state()

print('----------')
myobject.R2_move_contents_5L_to_3L()
myobject.print_state()

print('----------')
myobject.R1_fill_5L_jug()
myobject.print_state()

print('----------')
myobject.R2_move_contents_5L_to_3L()
myobject.print_state()

```
