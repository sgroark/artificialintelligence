Assignment - Problem 3: use DFS algorithm to return success if the goal is reachable or failure otherwise, given a start node and goal node.

```python
def DFS(graph, start_node, goal_node):

    stack = [start_node]
    visited = set()
    visited.add(start_node)
    path_found = False # boolean variable to determine if path was found

    parents = {start_node: start_node}

    while stack:
        vertex = stack.pop()

        # determines if goal node has been reached + returns path
        if vertex == goal_node:
            path_found = True
            reconst_path = []
            current_node = vertex

            while current_node != start_node:
                reconst_path.append(current_node)
                current_node = parents[current_node]

            # reverses the path order
            reconst_path.append(start_node)
            reconst_path.reverse()

            print('Path found: {}'.format(reconst_path))

        # visits child nodes
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)
                parents[neighbor] = vertex

    if not path_found:
            print('No path found.')
```
