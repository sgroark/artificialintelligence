```r
#install & load packages

install.packages('GA')
install.packages('ggplot2')
install.packages('plotly')

library(GA)
library(plotly)
```

```r
# function definitions

f1 <- function(x1,x2,x3,x4){
  return ((2*x2*x4) + (x3)*(x1 - 2*x4))
}

f2 <- function(x1,x2,x3,x4)
{
  return ((60000) / ((x3*(x1-2*x4)^3) + (2*x2*x4*((4*x4)^2 + 3*x1*(x1 - 2*x4)))))
}

#combined fitness function with 2 above functions as parameters

f_combined <- function(x, A, B)
{
  return ((A) * f1(x[1], x[2], x[3], x[4]) + (B) * f2(x[1], x[2], x[3], x[4]))
}

# upper and lower bounds for each parameter

x1 = seq(10,80, by = 0.1)
x2 = seq(10,50, by = 0.1)
x3 = seq(0.9,5.0, by = 0.1)
x4 = seq(0.9,5.0, by = 0.1)

lower <- c(10, 10, 0.9, 0.9)
upper <- c(80, 50, 5.0, 5.0)


# apply genetic algorithm

GA = ga(type = "real-valued",
        fitness = function(x) -f_combined(x, A = 0.5, B = 0.5),
        lower = lower, upper = upper,
        popSize = 50, maxiter = 100, pcrossover = 0.75, pmutation = 0.001, keepBest = TRUE
)

summary(GA)
plot(GA)


# print best solution from each generation

cat("\n\nList of best solutions from each generation: \n\n")
for (i in 1:GA@maxiter) {
  cat("Generation", i, "Solution:", GA@bestSol[[i]], "\n\n")

}
```
